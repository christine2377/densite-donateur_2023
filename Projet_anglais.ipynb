```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import fonctions as fct
from statsmodels.graphics.tsaplots import plot_acf
```

```python
AC=pd.read_csv("/content/tables_cours/AC.PA.csv",index_col = 0, parse_dates=True)
AC['log_close'] = np.log(AC.Close)
AC
```

```python
fig,ax = plt.subplots()
ln1 = ax.plot(AC.Close,c='r', label = "Cours de AC")
ax2 = ax.twinx()
ln2 = ax2.plot(AC.log_close, c='b', label= "Cours de AC en log")
lns= ln1+ln2
labels = [l.get_label() for l in lns]
ax.legend(lns,labels)
```

```python
plot_acf(AC.Close.value)
```
